## Datafile Paths and Configuration

```python
DATAFILE = '/Users/shams/querent-datascience/Diagnosis_ICD/data/icd_10_2017.csv'
COLUMNS = ['icd_code', 'xyz', 'short_desc', 'long_desc']
HEADERLINE = None
DOCUMENT_COLUMN = 'long_desc'
df_data = pd.DataFrame.from_csv(DATAFILE, header=HEADERLINE)
if len(COLUMNS):
    df_data.columns = COLUMNS
df_data.head()
```

```python
df_data[DOCUMENT_COLUMN] = df_data[DOCUMENT_COLUMN].apply(pre._text_preprocessing)
documents = df_data[DOCUMENT_COLUMN].tolist()
```

## Running TF-IDF Analysis

```python
vocab = list(set([i for sub_doc in documents for i in sub_doc.strip().split()]))
vectorizer = TfidfVectorizer(
    stop_words=None,
    norm=None,
    min_df=0,
    sublinear_tf=True,
    token_pattern=r'(?u)\b\w+\b',
    vocabulary=vocab
)
```

```python
def tf_idf(documents):
    return vectorizer.fit_transform(documents)
```

```python
x = tf_idf(documents=documents)
```

```python
tf_idf = pd.DataFrame({'word': vectorizer.vocabulary, 'idf': vectorizer.idf_}, index=None)
tf_idf.sort_values('idf')
```
