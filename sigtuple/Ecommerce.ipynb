```python
dict_hierarchy = {}
for elem in hierarchy:
    l1 = dict_hierarchy.get(elem[0], [])
    l1.append(elem[1])
    dict_hierarchy[elem[0]] = l1
```

```python
tree = {}
def update_create_node(val, parent, children):
    if val in tree:
        tree[val]['c'] += children
    else:
        tree[val] = {'p': parent, 'c': children}
```

```python
def create_tree(dict_hierarchy, current = 1, parent = None):
    update_create_node(current, parent, dict_hierarchy.get(current, []))
    for elem in dict_hierarchy.get(current, []):
        create_tree(dict_hierarchy, elem, current)
```

```python
dict_discount = {}
for elem in hierarchy:
    dict_discount[elem[0]] = 0
    dict_discount[elem[1]] = 0
```

```python
def update_keys(keys, d):
    for elem in keys:
        dict_discount[elem] = d
```

```python
def find_all_keys(keys):
    update = []
    find_further = keys
    while find_further:
        for elem in find_further:
            update.append(elem)
            find_further = find_further[1:] + dict_hierarchy.get(elem, [])
    return update
```

```python
x = find_all_keys([1334])
x.sort()
x
```

```python
for q in queries:
    if q[0].lower() == 'q':
        print(dict_discount[int(q[1])])
    if q[0].lower() == 'u':
        key = int(q[1])
        d = int(q[2])
        keys_found = find_all_keys([key])
        if len(keys_found):
            update_keys(keys_found, d)
            
```
